```python
import taxislib
```

```python
tree_1: taxislib.Tree[str] = taxislib.Tree()
root_node: taxislib.Node[str] = tree_1.append_data(node_data="*")

sub_1: taxislib.Node[str] = tree_1.append_data(node_data="+", parent=root_node.id)
tree_1.append_data(node_data="x1", parent=sub_1.id)
tree_1.append_data(node_data="x2", parent=sub_1.id)

sub_2: taxislib.Node[str] = tree_1.append_data(node_data="-", parent=root_node.id)
tree_1.append_data(node_data="x1", parent=sub_2.id)
tree_1.append_data(node_data="x2", parent=sub_2.id)

print(tree_1)
```
